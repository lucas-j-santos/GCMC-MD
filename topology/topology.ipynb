```python
import mbuild as mb
import foyer
from gmso import Topology
from gmso.external.convert_mbuild import to_mbuild
```

```python
framework = Topology.load('IRMOF-1_2x2x2.mol2')
framework.identify_connections()
framework = to_mbuild(framework)
framework.name = 'IRMOF-1'
framework.box = mb.Box(lengths=[2.5832*2, 2.5832*2, 2.5832*2], angles=[90.00, 90.00, 90.00])
```

```python
ff_mof = foyer.Forcefield('Dubbeldam2007FlexibleIRMOF-1.xml')
framework_ff = ff_mof.apply(framework, assert_improper_params=True)
framework_ff.save('box.gro', overwrite=True, combine='all')
framework_ff.save('box.top', overwrite=True, combine='all')
```

```python
import forcefield_utilities as ffutils
from gmso import Topology, Box
from gmso.parameterization import apply

framework = Topology.load('IRMOF-1.mol2')
framework.identify_connections()
framework.box = Box(lengths=[2.5832, 2.5832, 2.5832], angles=[90.00, 90.00, 90.00])
ff = ffutils.FoyerFFs().load('Dubbeldam2007FlexibleIRMOF-1.xml').to_gmso_ff()

apply(top=framework,
      forcefields=ff,
      identify_connections=True)
```

```python
framework.save('irmof-1.gro', overwrite=True)
framework.save('irmof-1.top', overwrite=True)
```
